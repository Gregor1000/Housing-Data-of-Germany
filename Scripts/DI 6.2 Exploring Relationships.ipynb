## 03. Correlation

```python
# creating correlation matrix
df.corr()
```

```python
# creating correlation heapmap with seaborn
f,ax = plt.subplots(figsize=(14,14))
corr = sns.heatmap(df.corr(), annot = True, ax = ax)
```

```python
corr.figure.savefig(os.path.join (path, '04 Analysis','Visualizations','corr.png'))
```

#### The most important variable in this dataset is the 'Price'. Therefore I will start with interpreting this heatmap regarding to this variable:
#### Interesstingly enough there is only one variable ('Living_space') that has a moderate to strong relationship. Then we have 'Rooms' and Bathrooms' with weak to moderate relationship. And the rest has only weak or no relationship. That there is no relationship between 'Price and 'Year_built' and 'Year_renovated' is probably owed to the fact that 'Furnishing_quality' and 'Condition' reflect more the quality of housing than the two dates. Also one can interpret that 'Living_space' encupsulates all the other variables in one go. Eg. if 'Living_space' is small there will not be 2 bathrooms in a house. But the basic interpretation is, that the more rooms and living space a house has the more expensive it gets.
#### The strongest correlation is between 'Rooms' and 'Living_space' and 'Bathrooms' and 'Bedrooms' and 'Rooms' and 'Bathrooms'. This meens that more living space will be divided in more rooms. And if there are more rooms, more are categorized as bedrooms and bathrooms
#### But I think it also important that variables that are no numbers should be analysed. Therefor I will create dummy variables (1, 2, 3 etc) for 'Furnishing_quality', 'Condition' and 'Energy_efficiency_class' to broaden one's view into the data. (I did add this new columns in the better suitable script of DI 6.01 Sourcing Open Data)
#### After analysing the correlation matrix with the new dummy variables, there is again the pattern, that 'furnish_dummy', 'condition_dummy' and 'energy_dummy' have a weak or no correletion with price. 
#### I understand why 'Living_Space' plays such a crutial role for the price, but I don't understand why the other variables play only a minor role. Maybe the relationship with these other variables is NOT linear.
